- `ROADMAP`, which displays the default road map view,

# **Get the Basemap from WMS**

```python
import geemap
peta = geemap.Map(center=[1,115], zoom=5)
naip_url = 'https://basemap.nationalmap.gov/arcgis/services/USGSImageryTopo/MapServer/WMSServer?'
peta.add_wms_layer(url=naip_url, layers='0', name='USGSImageryTopo', format='image/png', shown=True)
peta
```
